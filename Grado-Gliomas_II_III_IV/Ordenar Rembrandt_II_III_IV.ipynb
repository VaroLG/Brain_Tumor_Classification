```python
import os
import pydicom
import numpy as np
import cv2  # Usar OpenCV para guardar PNG

def dicom_to_png_with_folders(input_folder, output_folder):
    """
    Convierte imgenes DICOM a PNG organizadas en carpetas correspondientes a la estructura original.
    
    Args:
    - input_folder (str): Ruta de la carpeta que contiene las imgenes DICOM.
    - output_folder (str): Ruta de salida donde se guardarn las imgenes PNG organizadas.
    """
    for root, _, files in os.walk(input_folder):
        # Calcular ruta relativa para replicar la estructura
        relative_path = os.path.relpath(root, input_folder)
        output_subdir = os.path.join(output_folder, relative_path)

        if not os.path.exists(output_subdir):
            os.makedirs(output_subdir)

        for file in files:
            if file.endswith('.dcm'):
                dicom_path = os.path.join(root, file)
                dicom_data = pydicom.dcmread(dicom_path)
                img = dicom_data.pixel_array

                # Normalizar la imagen
                img = (img - np.min(img)) / (np.max(img) - np.min(img))
                img = (img * 255).astype(np.uint8)

                # Guardar como PNG en la carpeta correspondiente
                output_file = os.path.join(output_subdir, f"{os.path.splitext(file)[0]}.png")
                cv2.imwrite(output_file, img)
                print(f"Guardado: {output_file}")

# Directorios
input_dir = r"Directorio\de\entrada\Gliomas_I_II_III"  # Cambiar segn sea necesario
output_dir = r"Directorio\de\salida\Gliomas_I_II_III"  # Cambiar segn sea necesario

dicom_to_png_with_folders(input_dir, output_dir)
```

Una vez transformadas las imgenes de formato DICOM a formato png comprobamos la salida. Al estar ya clasificadas por carpetas en funcin del grado de glioma vamos a extraer las imgenes que se encuentran dentro de subcarpetas con cdigos de paciente a una careta donde solo haya imgenes de ese grado tumoral 

```python
import os
import shutil

def copiar_imagenes(ruta_entrada, ruta_salida, extensiones=("jpg", "png", "jpeg")):
    """
    Copia todas las imgenes de las subcarpetas de ruta_entrada a la carpeta ruta_salida.

    Args:
        ruta_entrada (str): Ruta de la carpeta raz donde buscar imgenes.
        ruta_salida (str): Ruta de la carpeta de destino donde guardar las imgenes.
        extensiones (tuple): Extensiones de imgenes a copiar (por defecto jpg, png, jpeg).
    """
    # Crear la carpeta de salida si no existe
    if not os.path.exists(ruta_salida):
        os.makedirs(ruta_salida)

    # Contador para imgenes copiadas
    contador = 0

    # Recorrer la ruta de entrada
    for raiz, _, archivos in os.walk(ruta_entrada):
        for archivo in archivos:
            # Verificar si el archivo tiene una de las extensiones vlidas
            if archivo.lower().endswith(extensiones):
                ruta_origen = os.path.join(raiz, archivo)
                ruta_destino = os.path.join(ruta_salida, archivo)

                # Asegurarse de no sobrescribir archivos existentes en la carpeta de salida
                base, extension = os.path.splitext(archivo)
                contador_copia = 1
                while os.path.exists(ruta_destino):
                    ruta_destino = os.path.join(ruta_salida, f"{base}_{contador_copia}{extension}")
                    contador_copia += 1

                # Copiar el archivo
                shutil.copy2(ruta_origen, ruta_destino)
                contador += 1

    print(f"Se copiaron {contador} imgenes a la carpeta: {ruta_salida}")

# Ejemplo de uso
ruta_entrada =  r"Ruuta\de\entrada\G_II"
ruta_salida = r"Ruta\de\salida\G_II"
copiar_imagenes(ruta_entrada, ruta_salida)
```
