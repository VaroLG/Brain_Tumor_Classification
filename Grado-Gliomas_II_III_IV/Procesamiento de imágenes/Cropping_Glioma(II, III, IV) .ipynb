```python
import tensorflow as tf
import numpy as np
import os
```

```python
dir = r"Directorio donde se encuentran las carpetas"
os.listdir(dir)
```

```python
input_folder = r"carpeta de imagenes G_II"  # Carpeta imágenes originales
output_folder = r"carpeta de imagenes recortadas G_II_cropped" # Carpeta imágenes recortadas

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

def crop_black_background(image):
    # Convertir a escala de grises
    grayscale = tf.image.rgb_to_grayscale(image)
    
    # Crear máscara para píxeles no negros
    binary_mask = tf.cast(grayscale > 20, tf.int32)  # Umbral más bajo para considerar "negro"
    
    # Coordenadas no negras
    coords = tf.where(binary_mask > 0)
    
    if tf.shape(coords)[0] == 0:  # Si no hay contenido, devolver la imagen original
        return image
    
    # Límites del recorte
    y_min = tf.reduce_min(coords[:, 0])
    y_max = tf.reduce_max(coords[:, 0])
    x_min = tf.reduce_min(coords[:, 1])
    x_max = tf.reduce_max(coords[:, 1])
    
    # Aplicar recorte
    cropped = image[y_min:y_max + 1, x_min:x_max + 1, :]
    return cropped

# Procesar imágenes
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)
        
        # Leer imagen
        image = tf.io.read_file(input_path)
        image = tf.image.decode_image(image, channels=3)
        
        # Recortar fondo negro
        cropped_image = crop_black_background(image)
        
        # Convertir a uint8 y guardar
        cropped_image = tf.cast(cropped_image, tf.uint8)
        encoded_image = tf.image.encode_png(cropped_image)
        tf.io.write_file(output_path, encoded_image)
        
        print(f"Procesada y guardada: {output_path}")

print("Recorte completado.")
```
